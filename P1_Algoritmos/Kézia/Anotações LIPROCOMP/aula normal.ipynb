```python
horario = 18
def PresentePraMae():
    if horario >= 18:
        print('Boa noite.')
    elif horario >= 12:
        print('Boa tarde')
    elif horario >= 6:
        print('Bom dia.')

PresentePraMae()

#se você declarar uma variável fora da função ela não vai funcionar, é preciso declarar dentro pra que rode.
```

```python
def PresentePraMae(horario):
    if horario >= 18:
        print('Boa noite.')
    elif horario >= 12:
        print('Boa tarde')
    elif horario >= 6:
        print('Bom dia.')

entrada = int(input('Insira a hora: '))
PresentePraMae(entrada)

#se você declarar uma variável fora da função ela não vai funcionar, é preciso declarar dentro pra que rode.
```

```python
def printarValores(*valores):
    # não existe restrinção de valores e eles não viram listas
    # se tirar o * ele vai se transformar em uma lista.
    for i in valores:
        print(i)
# não existe restrinção de valores
printarValores(1,2,3,4,5)
```
