5.	Escreva um programa que leia 10 números e informe o maior e o menor número.

```python
maiorNumero = 0
menorNumero = 0

for i in range(10):
    num = int(input("Num: "))
    if num > maiorNumero:
        maiorNumero = num
    elif num < menorNumero:
        menorNumero = num
        
print("Maior = ", maiorNumero) 
print("Menor = ", menorNumero) 
```

7.	Escreva um programa que calcula o fatorial de um dado número N.

4! = 4 x 3 x 2 x 1 

```python
num = int(input("Num: "))

if num < 0:
    print("Não existe fatorial")
else:
    fatorial = 1
    for i in range(num,0,-1):
        fatorial *= i # fatorial = fatorial * i 
    print(fatorial)
```

26.	Uma grande firma deseja saber qual é o empregado mais recente e qual é o mais antigo. Desenvolver um programa para ler um número indeterminado de informações contendo o número do empregado e o número de meses de trabalho deste empregado e imprimir o mais recente e o mais antigo. Obs.: A última informação contém os dois números iguais a zero. Não existem dois empregados admitidos no mesmo mês.

```python
numEmpregado = int(input("Numero empregado: "))
mesesEmpregado = int(input("Meses empregado: "))

#Variáveis auxiliares para identificação do empregado mais antigo e do mais recente
numEmpregadoMaisRecente = numEmpregadoMaisAntigo = numEmpregado
mesesEmpregadoMaisRecente = mesesEmpregadoMaisAntigo = mesesEmpregado

while numEmpregado!= 0 or mesesEmpregado!= 0:
    if mesesEmpregado < mesesEmpregadoMaisRecente:
        mesesEmpregadoMaisRecente = mesesEmpregado
        numEmpregadoMaisRecente = numEmpregado
    
    if mesesEmpregado > mesesEmpregadoMaisAntigo:
        mesesEmpregadoMaisAntigo = mesesEmpregado
        numEmpregadoMaisAntigo = numEmpregado 
    
    numEmpregado = int(input("Numero empregado: "))
    mesesEmpregado = int(input("Meses empregado: "))

print(numEmpregadoMaisRecente, numEmpregadoMaisAntigo)
```

17.Faça um programa que mostre os n termos da Série a seguir: 

  S = 1/1 + 2/3 + 3/5 + 4/7 + 5/9 + ... + n/m. 
Imprima no final a soma da série. 



```python
s = 0
denominador = 1

n = int(input())


for i in range(1,n+1):
    s += i/denominador # s = s + i/denominador
    print(i,"/",denominador)
    denominador += 2

```
