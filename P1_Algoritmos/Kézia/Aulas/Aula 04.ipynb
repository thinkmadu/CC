5.	Faça um Programa que peça o raio de um círculo, calcule e mostre sua área. 

```python
# versão 1 

pi = 3.14159

raio = float(input("Informe o raio: "))

area = pi * (raio ** 2)

print (area)
```

```python
# versão 2 -> Mais utilizada

import math


raio = float(input("Informe o raio: "))

area = math.pi * math.pow(raio,2)

print(area)

print('{:.3f}'.format(area))
```
