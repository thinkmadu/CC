```python
matriz = [(1,2,3,5), (2,4,5,9),(7,8,0,2), (6,3,4,0)]

def rotateLeft(matriz):
    b = [[matriz[j][i] for j in range(4)] for i in range(3, -1, -1)]
    return b
for i in range(len(matriz)):
    print(rotateLeft(matriz)[i])
print('========================')
for i in range(len(matriz)):
    print(matriz[i])

```
