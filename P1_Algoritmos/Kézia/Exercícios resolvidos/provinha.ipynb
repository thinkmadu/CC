1. (5.0) Escreva uma função em Python para solucionar o seguinte problema: dadas 2 listas com n números entre 0 e 9, interpretadas como dois números inteiros de n dígitos, calcular uma lista que representa a somatória dos dois números.

```python
lista1 = list(map(int, input("Digite os nº da 1ª lista: ").strip().split()))
lista2 = list(map(int, input("Digite os nº da 2ª lista: ").strip().split()))

def somaElementos(lista1,lista2):
    lista1.reverse()
    lista2.reverse()
    lista3 = list()
    transporte = 0
    
    lista1.append(0)
    lista2.append(0)
    
    while len(lista1) != len(lista2):
        if len(lista1) > len(lista2):
            lista2.append(0)
        
        else:
            lista1.append(0)

    for i in range (len(lista1)):
        soma = lista1[i] + lista2[i] + transporte
        transporte = soma // 10

        if soma > 9:
            novoNum = soma % 10
            lista3.append(novoNum)
        else:
            lista3.append(soma)
            
    lista3.reverse()
    return lista3

print(somaElementos(lista1,lista2))
```

A função deve apresentar uma mensagem informando se o determinante da matriz fornecida é ou não igual a zero segundo a propriedade acima.
