1. Leia uma matriz 4 x 4, conte e escreva quantos valores maiores que 10 ela possui. 

```python
matriz = [[10, 2, 30, 40],[15, -86, 1, 3],[19, 1, 1, 1],[5, 1, 7, 1]]
maioresQ10 = 0

# 1º ciclo vai analisar as linhas
for linhas in range(len(matriz)):
    # 2º ciclo vai analisar as colunas começando do elemento 0
    for colunas in range(len(matriz[0])):
        # nesse if eu tive que colocar +1 porque ele começa a contar do 0 e tava dando um valor a menos.
        if matriz[linhas][colunas] + 1 > 10:
            maioresQ10 += 1

print('Quantidade de nº maiores que 10: ',maioresQ10)
```
