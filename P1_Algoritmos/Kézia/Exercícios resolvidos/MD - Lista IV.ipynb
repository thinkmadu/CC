1. Escreva um programa que efetue a leitura de 20 valores numéricos e apresente no final o total do somatório e a média dos valores lidos

```python
soma = 0
n = 20
# aqui eu vou usar soma como variável auxiliar pra ficar melhor

for i in range(n):
    num = float(input("Digite sua nota: "))
    soma += + num
    # para i, o ciclo vai se repetir n(20) vezes.
    # num vai ser a entrada dos dados, soma vai ser a soma, normal
    # depois de repetir o ciclo 20 vezes, ele vai sair do loop e vai mostrar a media e a soma

media = soma/n
print(soma)
print(media)
```

23. O Departamento Estadual de Meteorologia lhe contratou para desenvolver um programa que leia um conjunto indeterminado de temperaturas, e informe ao final a menor e a maior temperatura informada, bem como a média das temperaturas. 

```python
x = True
cont = []
contador = 0

while x:
    temperatura = float(input('Digite uma temperatura: '))
    contador += + 1
    cont.append(temperatura)

    fator = int(input('Deseja adicionar mais alguma coisa? 1 - sim, 0 - não '))
    if fator == 0:
        x =  False
    else:
        x = True

soma = sum(cont)
maxima = max(cont)
minima = min(cont)
print('A média de temperatura é {}, com máxima de {} e mínima de {}'.format(soma,maxima,minima))
```

27. A prefeitura de uma cidade fez uma pesquisa entre seus habitantes, coletando dados sobre o salário e número de filhos. A prefeitura deseja saber:
a)Média do salário da população;
b)Média do número de filhos;
c)Maior salário;
d)Percentual de pessoas com salário até R$250,00.
Desenvolver um programa para calcular e escrever o que foi pedido nos itens a, b, c e d. O final da leitura de dados se dará com a entrada de um salário negativo.

```python
salarios = []
salarios250 = []
filhos = []
listaTOTAL = []
x = True

while x:
    salario = float(input('Digite o valor do salário: '))
    Qfilhos = int(input('Digite a quantidade de filhos: '))

    if salario < 0:
        print('Você inseriu um salário negativo. Too bad, boy.')
        break
    else:
        x = True
        filhos.append(Qfilhos)
        if salario <= 250:
            salarios250.append(salario)
        else:
            salarios.append(salario)
        
        cont = int(input('Deseja continuar inserindo dados? 1 - SIM | 0 - NÃO: '))
        if cont == 1:
            x = True
        else: 
            x = False

soma = sum(salarios) + sum(salarios250)
mediaSalarios = soma / (len(salarios) + len(salarios250))
mediaFilhos = sum(filhos) / len(filhos)
maxSalario = max(salarios)

print('MÉDIA SALÁRIO DA POPULAÇÃO: R${}'.format(mediaSalarios))
print('MÉDIA DE FILHOS: {:.0f}'.format(mediaFilhos))
print('MAIOR SALÁRIO DA PESQUISA: R${}'.format(maxSalario))

```
