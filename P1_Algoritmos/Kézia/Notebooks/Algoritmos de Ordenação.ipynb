```python
import random 

'''  Buble sort 
A cada passagem da sequencia o elemento com menor valor de acordo do tipo ordenação, deve ser flutuado para o topo
Cada elemento da posição i será comparado com o elemento da posição i + 1

'''

def bubbleSort(lista):
    n = len(lista)
    for i in range(n):
        # os últimos i elementos já estão no lugar
        for j in range(n-i-1):
            if lista[j] > lista[j+1] :
                lista[j], lista[j+1] = lista[j+1], lista[j]
    return lista
                

                
                      
lista = list(random.sample(range(-100, 100), 10))
print("Lista original: ", lista)
print("Lista ordenada: ", bubbleSort(lista))
```

```python
import random

''' Selection Sort

Para um vetor A de n elementos, execute os seguintes passos:

    1 - A cada iteração do algoritmo, selecione o menor elemento do vetor A
        e coloque-o em sua “posição correta”.
    2 - Resolva o problema para os n−1 elementos restantes.'''


def selectionSort(lista):
    n = len(lista)
    
    for i in range(n):
        # Encontra o elemento mínimo em A.
        minimo = i
        
        for j in range(i + 1, n):
            if lista[minimo] > lista[j]:
                minimo = j
                
        # Coloca o elemento mínimo na posição correta.
        lista[i], lista[minimo] = lista[minimo], lista[i]
    return  lista

        
lista = list(random.sample(range(-100, 100), 10))
print("Lista original: ", lista)
print("Lista ordenada: ", selectionSort(lista))
```

```python
import random


''' Insertion Sort


Para uma lista A de n elementos, execute os seguintes passos:

    A cada iteração i do algoritmo, troque de posição o elemento A[i] com cada elemento maior que
    ele que esteja à sua esquerda.
'''

''' A cada iteração i do algoritmo, todos os elementos à esquerda do índice i estarão ordenados e 
    elementos à direita do elemento no índice i ainda não foram processados (podem ser menores, 
    iguais ou maiores que o elemento em A[i]).'''


def insertionSort(lista):
    n = len(lista)
    
    for j in range(1, n):
        chave = lista[j]
        i = j - 1
      
        while i >= 0 and lista[i] > chave:
            lista[i + 1] = lista[i]
            i = i - 1
        
        lista[i + 1] = chave
    return lista


lista = random.sample(range(-100, 100), 10)
print("Lista original: ", lista)
print("Lista ordenada: ", insertionSort(lista))
```
