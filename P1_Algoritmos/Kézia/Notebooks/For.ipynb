4.	Fazer um programa para ler o sexo de dez pessoas e mostrar a quantidade de pessoas do sexo masculino e do sexo feminino, separadamente.

```python
contMasculino = contFeminino = 0

for i in range(3):
    sexo = input("Informar apenas m e f: ")
    
    if sexo == "m":
        contMasculino = contMasculino + 1 # contMasculino += 1
    elif sexo == "f":
        contFeminino += 1
    else:
        print("Sexo inválido")
        
print(contMasculino,contFeminino)
        
```
